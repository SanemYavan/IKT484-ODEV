```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_excel("MakineOgrenmesi.xlsx")
df.head()
```

```python
features = [
    "FAİZ ORANI(%)",
    "ÜLKELERE GÖRE KONUT FİYATLARI (%)",
    "EV SAHİPLİĞİ ORANI",
    "KAMU HARCAMALARI (MİLYAR)",
    "OTOMOBİL ÜRETİMİ (ADET/BİRİM)",
    "HER 1000 KİŞİYE DÜŞEN OTOMOBİL SAYISI",
    "OTOMOTİV İHRACATI (USD)",
    "ŞEHİRLEŞME ORANI",
    "OTOMOBİL SAHİPLİĞİ"
]
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[features])
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
df["Küme"] = kmeans.fit_predict(scaled_data)
```

```python
df[["ÜLKE", "Küme"]]
```

```python
df.groupby("Küme")[features].mean()
```

```python
plt.figure(figsize=(8,6))
colors = ['red' if c == 0 else 'blue' for c in df["Küme"]]
plt.scatter(df["HER 1000 KİŞİYE DÜŞEN OTOMOBİL SAYISI"],
            df["ÜLKELERE GÖRE KONUT FİYATLARI (%)"],
            c=colors)

plt.xlabel("Her 1000 Kişiye Düşen Otomobil Sayısı")
plt.ylabel("Konut Fiyat Artışı (%)")
plt.title("Kümelemeye Göre Ülkelerin Dağılımı")
plt.grid(True)
plt.show()
```
